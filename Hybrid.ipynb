```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn. preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('TRAIN.csv')
x = df.drop(columns = ['class'])
y = df['class']
```

```python
for column in x.columns:
    if x[column].dtype == type(object):
        labelencoder = LabelEncoder()
        x[column] = labelencoder.fit_transform(x[column])
        
```

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 0, stratify = y)
```

```python
print(X_train.shape)
sc = StandardScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
y_train = np.array(Y_train)
y_test = np.array(Y_test)
```

```python
train_x, val_x, train_y, val_y = train_test_split(x_train, y_train, test_size = 0.2, random_state = 0, stratify = y_train)
```

```python
print(train_x.shape, val_x.shape)
```

```python
columns = x.columns
```

```python
model2=GaussianNB()
model2.fit(train_x, train_y)
y_val_2=model2.predict(val_x)
y_val_2=pd.DataFrame(y_val_2)
y_test_2=model2.predict(x_test)
y_test_2=pd.DataFrame(y_test_2)
```

```python
model1=DecisionTreeClassifier()
model1.fit(train_x, train_y)
y_val_1=model1.predict(val_x)
y_val_1=pd.DataFrame(y_val_1)
y_test_1=model1.predict(x_test)
y_test_1=pd.DataFrame(y_test_1)
```

```python
model3=RandomForestRegressor(n_estimators=20, random_state=0)
model3.fit(train_x, train_y)
y_val_3=model3.predict(val_x)
y_val_3=pd.DataFrame(y_val_3)
y_test_3=model3.predict(x_test)
y_test_3=pd.DataFrame(y_test_3)
```

```python
val_input = pd.concat([pd.DataFrame(val_x, columns=columns), y_val_1, y_val_2, y_val_3], axis=1)
test_input = pd.concat([pd.DataFrame(x_test, columns=columns), y_test_1, y_test_2, y_test_3], axis=1)
val_input
```

```python
model=RandomForestClassifier(n_estimators=200)
model.fit(val_input, val_y)
```

```python
print((model.score(test_input, y_test))*100, "%")
```
